```python
# función para asignar categoría educacional
def categoria_educacion(valor):
    if valor <= 7.0:
        return 'Basica'
    elif valor <= 9.0:
        return 'Media'
    else:
        return 'Tecnico/Profesional'

df['Educacion'] = df['P37'].apply(categoria_educacion)

# función para asignar categoría etaria (edad)
def categoria_etaria(valor):
    if valor <= 29.0:
        return '18 a 29 años'
    elif valor <= 39.0:
        return '30 a 39 años'
    elif valor <= 49.0:
        return '40 a 49 años'
    elif valor <= 59.0:
        return '50 a 59 años'
    else:
        return '60 años o mas'

df['Edad'] = df['P35'].apply(categoria_etaria)
```
